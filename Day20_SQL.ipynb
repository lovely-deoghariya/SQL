```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display employees whose employee_id is BETWEEN 1 AND 10
pd.read_sql("""
SELECT *
FROM employees
WHERE employee_id BETWEEN 1 AND 10;
""", conn)
```

```python
#Display customers whose city NOT delhi
pd.read_sql("""
SELECT *
FROM customers
WHERE NOT city = 'Delhi';
""", conn)
```

```python
#Display products whose category is IN (electronic, furniture) AND price BETWEEN 1000 AND 5000
pd.read_sql("""
SELECT *
FROM products
WHERE category IN ('Electronics', 'Furniture')
AND price BETWEEN 1000 AND 5000;
""", conn)
```

```python
#Display orders where customer_id IN (1,2,3) AND employee_id NOT IN (4,5)
pd.read_sql("""
SELECT *
FROM orders
WHERE customer_id IN (1, 2, 3)
AND order_id NOT IN (4, 5);
""", conn)
```

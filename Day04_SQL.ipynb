```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display only the department_name column from the department table
pd.read_sql("""
SELECT department_name
FROM departments;
""", conn)
```

```python
#Display only first_name and email from the employees table
pd.read_sql("""
SELECT first_name, email
FROM employees;
""", conn)
```

```python
#Display product_name and price from the products table
pd.read_sql("""
SELECT product_name, price
FROM products;
""", conn)
```

```python
#Display customer_name and city from the customers table
pd.read_sql("SELECT * FROM customers LIMIT 5;", conn)
```

```python
#Display order_id, customer_id and order_date from the orders table
pd.read_sql("""
SELECT order_id, customer_id, order_date
FROM orders;
""", conn)
```

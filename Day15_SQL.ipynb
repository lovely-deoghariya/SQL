```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display customer whose city is IN (Bangalore, Mumbai)
pd.read_sql("""
SELECT *
FROM customers
WHERE city IN ('Bangalore', 'Mumbai');
""", conn)
```

```python
#Display products whose category is IN (electronics, furniture)
pd.read_sql("""
SELECT *
FROM products
WHERE category IN ('Electronics', 'Furniture');
""", conn)
```

```python
#Display employees whose department_id is IN (1,2)
pd.read_sql("""
SELECT *
FROM employees
WHERE department_id IN (1, 2);
""", conn)
```

```python
#Display orders whose customer_id is IN (1,2,3)
pd.read_sql("""
SELECT *
FROM orders
WHERE customer_id IN (1, 2, 3);
""", conn)
```

```python
#Display employee_id , first_name and last_name for employees whose department_id is IN (1,2)
pd.read_sql("""
SELECT employee_id, first_name, last_name
FROM employees
WHERE department_id IN (2, 3);
""", conn)
```

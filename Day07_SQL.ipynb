```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display employee_id and last_name from the 5 employees
pd.read_sql("""
SELECT employee_id, last_name
FROM employees
LIMIT 5;
""", conn)
```

```python
#Display department_id and department_name from the first 3 departments
pd.read_sql("""
SELECT department_id, department_name
FROM departments
LIMIT 3;
""", conn)
```

```python
#Display product_name, category and price from the first 3 departments
pd.read_sql("""
SELECT product_name, category, price
FROM products
LIMIT 3;
""", conn)
```

```python
#Display customer_name and email from the first 6 customers
pd.read_sql("PRAGMA table_info(customers);", conn)
```

```python
#Display order_id, order_date, customer_id and employee_id from the first 10 orders
pd.read_sql("PRAGMA table_info(orders);", conn)
```

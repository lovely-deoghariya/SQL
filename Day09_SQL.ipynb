```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display employee_id and first name from the first 6 employees
pd.read_sql("""
SELECT employee_id, first_name
FROM employees
LIMIT 6;
""", conn)
```

```python
#Display department_id and department_name from the first 5 departments
pd.read_sql("""
SELECT department_id, department_name
FROM departments
LIMIT 5;
""", conn)
```

```python
#Display product_name, category and price from the first 8 products
pd.read_sql("""
SELECT product_name, category, price
FROM products
LIMIT 8;
""", conn)
```

```python
#Display customer_id, customer_name and email from the first 10 customers
pd.read_sql("""
SELECT customer_id, first_name, last_name, email
FROM customers
LIMIT 10;
""", conn)
```

```python
#Display order_id, customer_id, employee_id and order_date from the first 15 ordepd.
pd.read_sql("""
SELECT order_id, customer_id, product_id, order_date
FROM orders
LIMIT 15;
""", conn)
```

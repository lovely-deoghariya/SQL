```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display all employees whose employee_id is BETWEEN 2 AND 6
pd.read_sql("""
SELECT *
FROM employees
WHERE employee_id BETWEEN 2 AND 6;
""", conn)
```

```python
#Display customers whose city is IN (delhi, Mumbai)
pd.read_sql("""
SELECT *
FROM customers
WHERE city IN ('Delhi', 'Mumbai');
""", conn)
```

```python
#Display products whose category is NOT furniture AND price BETWEEN 500 AND 5000
pd.read_sql("""
SELECT *
FROM products
WHERE NOT category = 'Furniture'
AND price BETWEEN 500 AND 5000;
""", conn)
```

```python
#Display order where customer_id IN (1,2,3) AND employee_id IN (1,2)
pd.read_sql("""
SELECT *
FROM orders
WHERE customer_id IN (1, 2, 3)
AND order_id IN (1, 2);
""", conn)
```

```python
#Display employee_id, first_name, last_name for employees whose department_id = 2 or department_id = 3 but employee_id is NOT BETWEEN 1 AND 2
pd.read_sql("""
SELECT employee_id, first_name, last_name
FROM employees
WHERE (department_id = 2 OR department_id = 3)
AND employee_id NOT BETWEEN 1 AND 2;
""", conn)
```

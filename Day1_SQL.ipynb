```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display all columns from the employees tables
pd.read_sql("SELECT * FROM employees;", conn)
```

```python
#Display all columns from the department table
pd.read_sql("SELECT * FROM departments;", conn)
```

```python
#Display only first_name and last_name from employees
pd.read_sql("SELECT first_name, last_name FROM employees;", conn)
```

```python
#Display the first 5 rows from the products table
pd.read_sql("SELECT * FROM products LIMIT 5;", conn)
```

```python
#Display the first 10 rows from the customers table
pd.read_sql("SELECT * FROM customers LIMIT 10;", conn)
```

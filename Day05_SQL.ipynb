```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display the first 3 rows from the department table
pd.read_sql("""
SELECT *
FROM departments
LIMIT 3;
""", conn)
```

```python
#Display first_name and last_name from the first 5 employees
pd.read_sql("""
SELECT first_name, last_name
FROM employees
LIMIT 5;
""", conn)
```

```python
#Display product_name and price from the first 8 products
pd.read_sql("""
SELECT product_name, price
FROM products
LIMIT 8;
""", conn)
```

```python
#Display customer_name and city from the first 10 customers
pd.read_sql("PRAGMA table_info(customers);", conn)
```

```python
#Display order_id, customer_id and order_date from the first 12 orders
pd.read_sql("""
SELECT order_id, customer_id, order_date
FROM orders
LIMIT 12;
""", conn)
```

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display only department_name from department
pd.read_sql("""
SELECT department_name
FROM departments;
""", conn)
```

```python
#Display only product_name and print product
pd.read_sql("""
SELECT product_name
FROM products;
""", conn)
```

```python
#Display all unique city values from customers
pd.read_sql("""
SELECT DISTINCT city
FROM customers;
""", conn)
```

```python
#Display all unique job_tittle values from employees
pd.read_sql("""
SELECT DISTINCT job_title
FROM employees;
""", conn)
```

```python
#Display only first_name, last_name and email from employees
pd.read_sql("""
SELECT first_name, last_name, email
FROM employees;
""", conn)
```

```python
#Display all unique product categories from the products table
pd.read_sql("""
SELECT DISTINCT category
FROM products;
""", conn)
```

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display the employee who employee_id is 1 AND price is greater than 2
pd.read_sql("""
SELECT *
FROM employees
WHERE employee_id = 1
AND department_id = 1;
""", conn)
```

```python
#Display products where category is Electronics AND price is greater than 1000
pd.read_sql("""
SELECT *
FROM products
WHERE category = 'Electronics'
AND price > 1000;
""", conn)
```

```python
#Display customers whose city "Bangalore" AND customer_id is greater than 2
pd.read_sql("""
SELECT *
FROM customers
WHERE city = 'Bangalore'
AND customer_id > 2;
""", conn)
```

```python
#Display order where customer_id is 1 AND employee_id is 2
pd.read_sql("""
SELECT *
FROM orders
WHERE customer_id = 1;
""", conn)
```

```python
#Display first_name, last_name and email of employees whose department_id is 2 AND employee_id is greater than 3
pd.read_sql("""
SELECT first_name, last_name, email
FROM employees
WHERE department_id = 2
AND employee_id > 3;
""", conn)
```

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("/content/drive/MyDrive/SQL/company.db")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)
```

```python
#Display only employee_id and email from the first 5 employees
pd.read_sql("""
SELECT employee_id, email
FROM employees
LIMIT 5;
""", conn)
```

```python
#Display department_name from the first 4 department
pd.read_sql("""
SELECT department_name
FROM departments
LIMIT 4;
""", conn)
```

```python
#Display product_id, product_name and category from the first 6 products
pd.read_sql("""
SELECT product_id, product_name, category
FROM products
LIMIT 6;
""", conn)
```

```python
#Display order_id, customer_id, employee_id and order_date from the first 12 orders
pd.read_sql("""
SELECT order_id, customer_id, order_date
FROM orders
LIMIT 12;
""", conn)
```

```python
#Display the first 10 rows from the products table showing product_name and price only
pd.read_sql("""
SELECT product_name, price
FROM products
LIMIT 10;
""", conn)
```
